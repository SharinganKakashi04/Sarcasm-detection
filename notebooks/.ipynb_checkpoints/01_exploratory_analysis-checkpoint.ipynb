```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from collections import Counter
import re

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Load data
print("Loading datasets...")
train_df = pd.read_csv('../data/train.csv')
val_df = pd.read_csv('../data/val.csv')
test_df = pd.read_csv('../data/test.csv')

print(f"Train: {len(train_df)} samples")
print(f"Val: {len(val_df)} samples")
print(f"Test: {len(test_df)} samples")

# ============================================
# 1. CLASS DISTRIBUTION
# ============================================
print("\n" + "="*50)
print("CLASS DISTRIBUTION")
print("="*50)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for idx, (df, name) in enumerate([(train_df, 'Train'), (val_df, 'Val'), (test_df, 'Test')]):
    counts = df['label'].value_counts()
    axes[idx].bar(['Non-Sarcastic', 'Sarcastic'], counts.values, color=['skyblue', 'coral'])
    axes[idx].set_title(f'{name} Set')
    axes[idx].set_ylabel('Count')
    for i, v in enumerate(counts.values):
        axes[idx].text(i, v + 50, str(v), ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('../results/class_distribution.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"\nTrain - Sarcastic: {train_df['label'].sum()} ({train_df['label'].mean()*100:.1f}%)")
print(f"Val - Sarcastic: {val_df['label'].sum()} ({val_df['label'].mean()*100:.1f}%)")
print(f"Test - Sarcastic: {test_df['label'].sum()} ({test_df['label'].mean()*100:.1f}%)")

# ============================================
# 2. TEXT LENGTH ANALYSIS
# ============================================
print("\n" + "="*50)
print("TEXT LENGTH ANALYSIS")
print("="*50)

train_df['text_length'] = train_df['text'].str.len()
train_df['word_count'] = train_df['text'].str.split().str.len()

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Character length
for label in [0, 1]:
    data = train_df[train_df['label'] == label]['text_length']
    axes[0].hist(data, bins=50, alpha=0.6, label='Sarcastic' if label else 'Non-Sarcastic')
axes[0].set_xlabel('Character Length')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Text Length Distribution')
axes[0].legend()

# Word count
for label in [0, 1]:
    data = train_df[train_df['label'] == label]['word_count']
    axes[1].hist(data, bins=50, alpha=0.6, label='Sarcastic' if label else 'Non-Sarcastic')
axes[1].set_xlabel('Word Count')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Word Count Distribution')
axes[1].legend()

plt.tight_layout()
plt.savefig('../results/text_length_analysis.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nAverage lengths:")
print(f"Sarcastic - Chars: {train_df[train_df['label']==1]['text_length'].mean():.1f}, Words: {train_df[train_df['label']==1]['word_count'].mean():.1f}")
print(f"Non-Sarcastic - Chars: {train_df[train_df['label']==0]['text_length'].mean():.1f}, Words: {train_df[train_df['label']==0]['word_count'].mean():.1f}")

# ============================================
# 3. SENTIMENT ANALYSIS (VADER)
# ============================================
print("\n" + "="*50)
print("SENTIMENT INCONGRUITY ANALYSIS")
print("="*50)

analyzer = SentimentIntensityAnalyzer()

def get_vader_sentiment(text):
    scores = analyzer.polarity_scores(text)
    return scores['compound']

print("Computing VADER sentiment scores...")
train_df['vader_sentiment'] = train_df['text'].apply(get_vader_sentiment)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Distribution of VADER scores
for label in [0, 1]:
    data = train_df[train_df['label'] == label]['vader_sentiment']
    axes[0].hist(data, bins=50, alpha=0.6, label='Sarcastic' if label else 'Non-Sarcastic')
axes[0].set_xlabel('VADER Sentiment Score')
axes[0].set_ylabel('Frequency')
axes[0].set_title('VADER Sentiment Distribution')
axes[0].legend()
axes[0].axvline(x=0, color='black', linestyle='--', alpha=0.3)

# Box plot
train_df.boxplot(column='vader_sentiment', by='label', ax=axes[1])
axes[1].set_xlabel('Label (0=Non-Sarcastic, 1=Sarcastic)')
axes[1].set_ylabel('VADER Sentiment Score')
axes[1].set_title('VADER Scores by Class')
plt.suptitle('')

plt.tight_layout()
plt.savefig('../results/sentiment_analysis.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nAverage VADER scores:")
print(f"Sarcastic: {train_df[train_df['label']==1]['vader_sentiment'].mean():.3f}")
print(f"Non-Sarcastic: {train_df[train_df['label']==0]['vader_sentiment'].mean():.3f}")

# ============================================
# 4. LINGUISTIC FEATURES
# ============================================
print("\n" + "="*50)
print("LINGUISTIC FEATURES")
print("="*50)

def count_exclamations(text):
    return text.count('!')

def count_questions(text):
    return text.count('?')

def count_caps_words(text):
    words = text.split()
    return sum(1 for word in words if word.isupper() and len(word) > 1)

def has_ellipsis(text):
    return 1 if '...' in text or '' in text else 0

train_df['exclamations'] = train_df['text'].apply(count_exclamations)
train_df['questions'] = train_df['text'].apply(count_questions)
train_df['caps_words'] = train_df['text'].apply(count_caps_words)
train_df['ellipsis'] = train_df['text'].apply(has_ellipsis)

features = ['exclamations', 'questions', 'caps_words', 'ellipsis']

fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.ravel()

for idx, feature in enumerate(features):
    sarcastic = train_df[train_df['label']==1][feature].mean()
    non_sarcastic = train_df[train_df['label']==0][feature].mean()
    
    axes[idx].bar(['Non-Sarcastic', 'Sarcastic'], [non_sarcastic, sarcastic], 
                  color=['skyblue', 'coral'])
    axes[idx].set_title(f'Average {feature.replace("_", " ").title()}')
    axes[idx].set_ylabel('Average Count')
    
    for i, v in enumerate([non_sarcastic, sarcastic]):
        axes[idx].text(i, v, f'{v:.3f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.savefig('../results/linguistic_features.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nFeature averages:")
for feature in features:
    sarc_avg = train_df[train_df['label']==1][feature].mean()
    non_sarc_avg = train_df[train_df['label']==0][feature].mean()
    print(f"{feature}: Sarcastic={sarc_avg:.3f}, Non-Sarcastic={non_sarc_avg:.3f}")

# ============================================
# 5. SAMPLE EXAMPLES
# ============================================
print("\n" + "="*50)
print("SAMPLE EXAMPLES")
print("="*50)

print("\n SARCASTIC EXAMPLES:")
print("-" * 50)
sarcastic_samples = train_df[train_df['label']==1].sample(5, random_state=42)
for idx, row in sarcastic_samples.iterrows():
    print(f"Text: {row['text']}")
    print(f"VADER: {row['vader_sentiment']:.3f}\n")

print("\n NON-SARCASTIC EXAMPLES:")
print("-" * 50)
non_sarcastic_samples = train_df[train_df['label']==0].sample(5, random_state=42)
for idx, row in non_sarcastic_samples.iterrows():
    print(f"Text: {row['text']}")
    print(f"VADER: {row['vader_sentiment']:.3f}\n")

# ============================================
# 6. KEY INSIGHTS
# ============================================
print("\n" + "="*50)
print("KEY INSIGHTS FOR MODELING")
print("="*50)

print("\n1. Class Imbalance:")
imbalance_ratio = train_df['label'].value_counts()[0] / train_df['label'].value_counts()[1]
print(f"   Ratio (non-sarcastic:sarcastic) = {imbalance_ratio:.2f}:1")
print(f"    Consider: class weights, oversampling, or focal loss")

print("\n2. Sentiment Incongruity:")
sarc_vader = train_df[train_df['label']==1]['vader_sentiment'].mean()
non_sarc_vader = train_df[train_df['label']==0]['vader_sentiment'].mean()
print(f"   Sarcastic posts have {'more positive' if sarc_vader > non_sarc_vader else 'more negative'} VADER scores")
print(f"    This supports our incongruity hypothesis!")

print("\n3. Linguistic Markers:")
print(f"   Sarcastic posts use more: {', '.join([f for f in features if train_df[train_df['label']==1][f].mean() > train_df[train_df['label']==0][f].mean()])}")
print(f"    These features can help our model")

print("\n4. Text Length:")
print(f"   Average word count difference: {abs(train_df[train_df['label']==1]['word_count'].mean() - train_df[train_df['label']==0]['word_count'].mean()):.1f} words")

print("\n EDA Complete! Check the results/ folder for visualizations.")
```
